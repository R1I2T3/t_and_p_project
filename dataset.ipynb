```python
import pandas as pd

# Define category names or values
Category_1 = 'Category 1'
Category_2 = 'Category 2'
Category_3 = 'Category 3'
Category_4 = 'Category 4'

# Read the Excel file
df = pd.read_excel('/content/placement_data.xlsx')

# Apply the lambda function to categorize
df['CATEGORY'] = df.apply(
    lambda row: Category_1 if row['CGPA'] >= 7.84 and row['AY'] >= 75 and row['TY'] >= 75 and row['TEST SCORE'] >= 75 else
                Category_2 if row['CGPA'] >= 6.77 and row['AY'] >= 75 and row['TY'] >= 65 and row['TEST SCORE'] >= 65 else
                Category_3 if row['CGPA'] < 6.77 and row['AY'] >= 75 and row['TY'] < 65 and row['TEST SCORE'] < 65 else
                Category_4, axis=1)

# Print the DataFrame to verify
print(df)

# Save the updated DataFrame to a new Excel file
df.to_excel('updated_file.xlsx', index=False)
```
